```python
from scipy.stats import norm
```

```python
# Obtendo os dados de rhum_avg (umidade relativa média)

rhum_avg = df_agg['rhum_avg'].dropna()

# Média e desvio padrão da variável
media = rhum_avg.mean()
desvio = rhum_avg.std()

print(f"Média da umidade relativa média: {media:.2f}")
print(f"Desvio padrão: {desvio:.2f}")
```

```python
# Funções de distribuição acumulada para cada faixa

# Faixa 1: Aviso de Atenção (30% a 20%)
prob_30_20 = norm.cdf(30, media, desvio) - norm.cdf(20, media, desvio)

# Faixa 2: Alerta (20% a 12%)
prob_20_12 = norm.cdf(20, media, desvio) - norm.cdf(12, media, desvio)

# Faixa 3: Emergência (abaixo de 12%)
prob_12 = norm.cdf(12, media, desvio)
```

```python
# Impressão dos resultados em porcentagem

print(f"Probabilidade de Aviso de Atenção (30% a 20%): {prob_30_20:.2%}")
print(f"Probabilidade de Alerta (20% a 12%): {prob_20_12:.2%}")
print(f"Probabilidade de Emergência (abaixo de 12%): {prob_12:.2%}")
```

### Com a distribuição de Bernoulli, esperamos cerca de 10 dias com precipitação em 100 dias.

### Com a distribuição normal (FDA) aplicada à umidade relativa média, os riscos de condições críticas são muito baixos, sendo inferior a 0,2% para qualquer uma das faixas de atenção, alerta ou emergência.

## Questão 5

Hipotéticamente será necessário desenvolver uma Regressão para entender o comportamento de número de passeios que ocorreu no mês de Junho de 2018.

No papel de de um Ciêntista de Dados, selecione um tipo de Regressão ideal para a distribuição abaixo e justifique a escolha.


```python
# Criar
df_questao_5 = df_agg[['ride_date','rides']].copy()

# Mudar a tipagem da coluna ride_date de string para datetime64s
df_questao_5['ride_date'] = pd.to_datetime(df_questao_5['ride_date'])

# Filtro de data
df_questao_5 = df_questao_5[(df_questao_5['ride_date'] >= '2018-06-01') & (df_questao_5['ride_date'] <= '2018-06-30')]

# Gráfico de pontos
sns.scatterplot(data=df_questao_5, x='ride_date', y='rides')
plt.xticks(rotation=45)
plt.show()
```

# Resposta

À regressão ideal é Regressão Polinomial de Segunda Ordem (grau 2), para modelar o comportamento do número de passeios no mês de junho de 2018

O gráfico mostra um comportamento não linear ao longo do tempo. Podemos observar um declínio e elevação subsequente nos dados, formando um padrão semelhante a uma parábola ou curva suave. Isso indica que uma regressão linear simples (reta) não modelaria bem os dados, pois ela não conseguiria capturar essas oscilações.

Não há mudanças abruptas que justifiquem modelos mais complexos (como regressão spline ou árvores).

Uma regressão polinomial de grau 2 é simples, interpretável e suficiente para representar esse padrão observado (ex: subida, descida, subida).

A curva de tendência obtida mostra uma queda gradual na quantidade de passeios ao longo do mês, o que pode estar relacionado a fatores climáticos, sociais ou operacionais. Esse modelo permite antecipar padrões e apoiar decisões operacionais como o planejamento de demanda e alocação de recursos.



```python
# Importando o modelo de regressão linear

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Filtro de datas para Junho/2018

df_junho = df_questao_5[(df_questao_5['ride_date'] >= '2018-06-01') & (df_questao_5['ride_date'] <= '2018-06-30')]

# Converter datas para números (dias desde o início)
df_junho['dias'] = (df_junho['ride_date'] - pd.to_datetime('2018-06-01')).dt.days
```

```python
# Regressão Polinomial de 2º grau

X = df_junho[['dias']]
y = df_junho['rides']

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
```

```python
# Previsão para o gráfico

dias_plot = np.arange(0, 30).reshape(-1, 1)
dias_plot_poly = poly.transform(dias_plot)
y_pred = model.predict(dias_plot_poly)
```

```python
# Plotando os dados reais e a curva de regressão

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_junho['ride_date'], y=df_junho['rides'], label='Dados Reais')
plt.plot(pd.date_range(start='2018-06-01', periods=30), y_pred, color='red', label='Regressão Polinomial (grau 2)')
plt.xticks(rotation=45)
plt.title('Regressão Polinomial - Junho 2018')
plt.xlabel('Data')
plt.ylabel('Número de Passeios')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Questão 6

Formule a seguinte Hipótese: A empresa que gerencia a operação de bikes precisa saber de forma factual a necessidade de criar e investir em uma força tarefa para véspora de feriados, mas primeiramente, temos comprovar que essas datas incorre em aumento da utilização das bikes.

**Hipótese Nula (H0)**: A média de passeios do dataset df_agg é (μ=617).

**Hipótese Alternativa (H1)**: A média de passeios em feriados é superior (μ/=?).

Portanto, identifique os feriados que ocorream entre 2018-01-28 e 2018-08-30 calculando a μ da hípotese alternativa para obter o p-value para validação ou invalidação das hipóteses.

**Detalhe o Teste de Hipótese conforme feito em sala de aula.**

```python
df_agg['rides'].describe()
```

# Resposta

```python
# Função para fazer testes estatísticos

from scipy.stats import ttest_1samp
```

```python
# Transformar a coluna de datas para datetime

df_agg['ride_date'] = pd.to_datetime(df_agg['ride_date'])

# Lista de feriados nacionais e locais (Brasília)
feriados = [
    '2018-02-12',  # Carnaval
    '2018-02-13',  # Carnaval
    '2018-02-14',  # Carnaval
    '2018-03-30',  # Sexta-feira Santa
    '2018-04-21',  # Tiradentes
    '2018-05-01',  # Dia do Trabalho
    '2018-05-31',  # Corpus Christi
    '2018-06-01',  # Corpus Christi
    '2018-08-15'   # Dia de Nossa Senhora da Assunção (feriado em Brasília)
]

feriados = pd.to_datetime(feriados)
```

```python
# Filtrar os dados apenas nos dias de feriado

df_feriados = df_agg[df_agg['ride_date'].isin(feriados)]

# Hipótese nula: média de rides nos feriados = 617
media_populacional = 617

# Cálculo da média nos feriados
media_feriados = df_feriados['rides'].mean()

# Teste T de uma amostra
t_stat, p_value = ttest_1samp(df_feriados['rides'], popmean=media_populacional)

# Ajuste para teste unilateral (H1: média dos feriados > média populacional)
if media_feriados > media_populacional:
    p_value /= 2
else:
    p_value = 1 - (p_value / 2)
```

```python
# Resultados do teste

print("Média de rides nos feriados:", round(media_feriados, 2))
print("Média populacional (H0):", media_populacional)
print("Estatística t:", round(t_stat, 4))
print("p-valor (unilateral):", round(p_value, 6))

# Conclusão
alpha = 0.05
if p_value < alpha:
    print("\nRejeitamos H0. A média de passeios nos feriados é significativamente maior.")
else:
    print("\nNão rejeitamos H0. Não há evidência significativa de que a média nos feriados é maior.")
```

Para investigar se o número médio de passeios nos feriados foi estatisticamente maior do que a média geral de 617 passeios por dia, foi realizado um teste t para uma amostra (one-sample t-test) com os seguintes dados:

* Média populacional (H₀): 617 passeios por dia.

* Média observada nos feriados: 604,63 passeios.

* t-statistic: -0,070

* p-valor unilateral: 0,527

Como o p-valor unilateral foi maior que 0,05 (nível de significância adotado), não há evidências estatísticas suficientes para rejeitar a hipótese nula (H₀). Isso significa que não podemos afirmar que a média de passeios nos feriados foi maior do que a média geral, fazendo assim não rejeitarmos H₀.

A média de passeios nos feriados não foi significativamente maior do que a média geral de 617 passeios por dia.
